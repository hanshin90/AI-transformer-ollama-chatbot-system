#### ChatOllama 경로 체크

```python
import subprocess

ollama_path = r"C:\Users\Administrator\AppData\Local\Programs\Ollama\ollama.exe"

result = subprocess.run(
    [ollama_path, "--version"],
    capture_output=True,
    text=True
)

print("returncode:", result.returncode)
print("stdout:", result.stdout)
print("stderr:", result.stderr)
```
